```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
pymac = pd.merge(study_results, mouse_metadata, on="Mouse ID")

# Display the data table for preview
pymac
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

drugs = sorted(pymac['Drug Regimen'].unique())
counts = pymac.groupby('Drug Regimen')['Mouse ID'].count()
means = pymac.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
medians = pymac.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variances = pymac.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
stdevs = pymac.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sems = pymac.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
drugs
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
data = {
    'Mean Tumor Volume': means,
    'Median Tumor Volume': medians,
    'Tumor Volume Variance': variances,
    'Tumor Volume Std. Dev.': stdevs,
    'Tumor Volume Std. Err.': sems
}

summary_df = pd.DataFrame(data)
summary_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_df2 = pymac.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(
    Mean=np.mean,
    Median=np.median,
    Variance=np.var,
    Std_Dev=np.std,
    SEM=lambda x: np.std(x, ddof=1) / np.sqrt(len(x))
)
summary_df2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
final_tps = pymac.groupby('Mouse ID')['Timepoint'].last()
final_tps

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
# aka - use inner merge to select only the last timepoint rows for each mouse (1880 > 248)
pymac_final = pd.merge(pymac, final_tps, how='inner', on=["Mouse ID", 'Timepoint'])
pymac_final

# filter df for only rows with the 4 drugs of interest (248 > 100)
filtered_df = pymac_final[pymac_final['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])]
filtered_df
```

```python
# test / thought process step
# goal: find the IQR and outliers of the tumor vol column, for the 4 drugs, grouped by 

capomulin = pymac_final.loc[pymac_final['Drug Regimen'] == 'Capomulin', 'Tumor Volume (mm3)'].tolist()
capomulin
```

```python
# generate list of tumor volumes for the drugs of interest by creating a dict using loop

# define drugs of interest
drugs = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# create empty dict to store the results
results = {}

# loop through each drug
for drug in drugs:
    
    # filter df for only rows with that drug
    filtered_data = pymac_final[pymac_final['Drug Regimen'] == drug]
    
    # output the relevant tumor volumes as a list
    tumor_volume_list = filtered_data['Tumor Volume (mm3)'].tolist()
    
    # store the results in the dictionary
    results[drug] = tumor_volume_list

# print the results for each drug
for drug, values in results.items():
    print(f"Results for {drug}: {values}")
```

```python
results['Capomulin']
```

```python
# determine possible outliers

# copied drug list for visibility 
drugs = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# create dict to store outlier results
outlier_results = {}

# loop through each drug of interest
for drug in drugs:
    
    # get the tumor volume values for the current drug from the 'results' dictionary
    tumor_volume_list = results[drug]
    
    # calc the quartiles and iqr for the current drug's tumor volumes
    quartiles = pd.Series(tumor_volume_list).quantile([0.25, 0.5, 0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq

    # calc the lower and upper bounds for potential outliers
    lower_bound = lowerq - 1.5 * iqr
    upper_bound = upperq + 1.5 * iqr

    # dtm potential outliers for current drug
    potential_outliers = [value for value in tumor_volume_list if (value < lower_bound) or (value > upper_bound)]

    # log results
    outlier_results[drug] = {
        "Lower Quartile": lowerq,
        "Median": quartiles[0.5],
        "Upper Quartile": upperq,
        "IQR": iqr,
        "Lower Bound": lower_bound,
        "Upper Bound": upper_bound,
        "Potential Outliers": potential_outliers
    }

# print the results for each drug
for drug, values in outlier_results.items():
    print(f"Results for {drug}:")
    for key, value in values.items():
        print(f"  {key}: {value}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
data1 = (results['Capomulin'])
data2 = (results['Ramicane'])
data3 = (results['Infubinol'])
data4 = (results['Ceftamin'])
plt.boxplot([data1, data2, data3, data4], flierprops={'markerfacecolor': 'red', 'marker': 'o', 'markersize': 12})
plt.ylabel("Final Tumor Volume (mm3)")

custom_ticks = [1, 2, 3, 4]
plt.xticks(custom_ticks, drugs)
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

from scipy.stats import linregress

# copy data creation lines for visibility:
capo = pymac.loc[pymac['Drug Regimen'] == 'Capomulin']
capo_weights = capo.groupby('Mouse ID')['Weight (g)'].mean().tolist()
capo_tumors = capo.groupby('Mouse ID')['Tumor Volume (mm3)'].mean().tolist()

# define x and y values
x_values = capo_weights
y_values = capo_tumors

# create chart
plt.scatter(x_values,y_values, c='blue')
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title('Capomulin: Mouse Weight vs. Avg. Tumor Volume')

# linear regression calculations
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = np.array(x_values) * slope + intercept

# graph the best fit line
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,36),fontsize=15,color="red")

print(f"The correlation is: {rvalue}")
```
